I'm going to analyze data about the movies.

Movie_metadata.csv has information about movie titles, directors, actors, ratings, budgets, etc.

I wonder if spending a lot of budget will increase consumer ratings.

I drew a scatterplot as seen below, and I can't figure out the correlation because there are a few movies with a huge budget.



```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('movie_metadata.csv')
df.plot(kind='scatter', x='budget', y='imdb_score')
```

I think we need to remove some movies with a very big budget and analyze them again.

So, I'm going to remove movies with a budget larger than 75% point of budget plus 5 IQR and draw a scatterplot again.

```python
#Let's say 25% points in q1, 75% points in q3, and use these two values to get IQR(Interquartile Range).

q1 = df['budget'].quantile(0.25)
q3 = df['budget'].quantile(0.75)
iqr = q3 - q1

# Remove the outliers
condition = (df['budget'] > q3 + 5 * iqr)
df.drop(df[condition].index, inplace=True)

# draw scatter plot
df.plot(kind='scatter', x= 'budget', y='imdb_score')
```

As I drew it on the graph, spending a lot of budget doesn't always mean that it has a high rating.

Sometimes movies with lower budget have higher ratings.

If you modify the 'q3 + 5 * iqr' part, you can remove the outliers in more different formats and see the different results.

For example, if you modify it to q3 + 10 * iqr, you can delete less data and see the results.

I think it is a good exercise to try changing the conditions of this point by yourself and see what the best result is.
