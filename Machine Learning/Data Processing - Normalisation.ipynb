I'm going to use 'preprocessing' module in the sklearn library to normalise the data.

```python
from sklearn import preprocessing
import pandas as pd

PATIENT_FILE_PATH = 'liver_patient_data.csv'
pd.set_option('display.float_format', lambda x: '%.5f' % x) # Outputs up to 5 decimal places

liver_patients_df = pd.read_csv(PATIENT_FILE_PATH)
liver_patients_df
```

```python
# Names of the columns that I want to normalise.
features_df = liver_patients_df[['Total_Bilirubin','Direct_Bilirubin', 'Alkaline_Phosphotase', 'Alamine_Aminotransferase']]
features_to_normalize = ['Total_Bilirubin','Direct_Bilirubin', 'Alkaline_Phosphotase', 'Alamine_Aminotransferase']
```

```python
scaler = preprocessing.MinMaxScaler() # Use the 'scaler' to normalise the data.
normalized_data = scaler.fit_transform(features_df)

normalized_df = pd.DataFrame(normalized_data, columns = features_to_normalize)
 

# Test
normalized_df.describe()
```
